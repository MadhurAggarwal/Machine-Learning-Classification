```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
x=np.arange(-20,20,3)
y=sigmoid(x)
print(np.c_[x,y])
```

```python
# Plot z vs sigmoid(z)
fig,ax = plt.subplots(1,1,figsize=(5,3))
ax.plot(x, y, color="blue")
ax.set_title("Sigmoid function")
ax.set_ylabel('sigmoid(z)')
ax.set_xlabel('z')
draw_vthresh(ax,0)
```
